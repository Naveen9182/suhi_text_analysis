## Filling missing values

```python
final_df = pd.read_excel('/content/final_df.xlsx')
```

```python
final_df_subset = final_df[['age', 'Hispanic', 'new_patient', 'visit_number', 'day_readmit', 'sex_Male', 'AfricanAmerican', 'White',
                  'Additional_race', 'language_English', 'language_Spanish', 'language_Other']]
```

```python
final_df_subset.info()
```

```python
final_df_subset.dropna(subset=['day_readmit'], inplace=True)
```

```python
final_df_subset.info()
```

```python
final_df_subset.isna().sum(axis=1).value_counts()
```

```python
final_df_subset_with_na = final_df_subset[final_df_subset.isna().any(axis=1)]
final_df_subset_without_na = final_df_subset[~final_df_subset.isna().any(axis=1)]
```

```python
final_df_subset_with_na['day_readmit'].value_counts()
```

```python
final_df_subset_without_na['day_readmit'].value_counts()
```

```python
final_df_subset_without_na
```

```python
features_to_be_filled = ['Hispanic', 'new_patient', 'sex_Male']
for feature in features_to_be_filled:
  print(f"Filling missing values for {feature}")
  x_train, x_test, y_train, y_test = train_test_split(final_df_subset_without_na.drop(feature, axis=1), final_df_subset_without_na[feature], test_size=0.2, random_state=42)
  # Train a model to predict sex
  model = DecisionTreeClassifier()
  model.fit(x_train, y_train)
  y_pred = model.predict(x_test)
  accuracy = accuracy_score(y_test, y_pred)
  print(f"Accuracy: {accuracy}")
  print(f"Adjusted Mutual Info Score: {adjusted_mutual_info_score(y_test, y_pred)}")
```

```python
features_to_be_filled = ['age']
for feature in features_to_be_filled:
  print(f"Filling missing values for {feature}")
  x_train, x_test, y_train, y_test = train_test_split(final_df_subset_without_na.drop(feature, axis=1), final_df_subset_without_na[feature], test_size=0.2, random_state=42)
  # Train a model to predict sex
  model = DecisionTreeRegressor()
  model.fit(x_train, y_train)
  y_pred = model.predict(x_test)
  # Evaluate R-squared and Adjusted R-squared
  r2 = model.score(x_test, y_test)
  print(f"R-squared: {r2}")
  # Calculate Mean Squared Error
  mse = mean_squared_error(y_test, y_pred)
  print(f"Mean Squared Error: {mse}")
  print(f"Adjusted Mutual Info Score: {adjusted_mutual_info_score(y_test, y_pred)}")
```

# Min-Max Scaler on Age

```python

scaler = MinMaxScaler()
features_to_be_filled = ['Hispanic', 'new_patient', 'sex_Male']
for feature in features_to_be_filled:
  print(f"Filling missing values for {feature}")
  x_train, x_test, y_train, y_test = train_test_split(final_df_subset_without_na.drop(feature, axis=1), final_df_subset_without_na[feature], test_size=0.2, random_state=42)
  x_train['age'] = scaler.fit_transform(x_train[['age']])
  x_test['age'] = scaler.transform(x_test[['age']])
  # Train a model to predict sex
  model = KNeighborsClassifier()
  model.fit(x_train, y_train)
  y_pred = model.predict(x_test)
  accuracy = accuracy_score(y_test, y_pred)
  print(f"Accuracy: {accuracy}")
  print(f"Adjusted Mutual Info Score: {adjusted_mutual_info_score(y_test, y_pred)}")
```

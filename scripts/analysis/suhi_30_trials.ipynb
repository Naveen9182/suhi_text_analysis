```python
g = suhi_30_trial_df.groupby(GROUP_COLS)[NUMERIC_COL]

summary = g.agg(
    mean      = "mean",
    std       = "std",
    n         = "count"
).reset_index()

summary["se"]             = summary["std"] / np.sqrt(summary["n"])
summary["half_ci"]        = 1.96 * summary["se"]               # or Student-t if you like
summary["ci_low"]         = summary["mean"] - summary["half_ci"]
summary["ci_high"]        = summary["mean"] + summary["half_ci"]
summary["mean ± 95% CI"]  = summary["mean"].round(2).astype(str) \
                           + " ± " \
                           + summary["half_ci"].round(2).astype(str)

print(summary)
```

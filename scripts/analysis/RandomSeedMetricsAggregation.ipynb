```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
training_log_df = pd.read_csv('/content/training_log_05_14.csv')
```

```python
training_log_df.drop(['Run ID', 'Random Seed', 'File Name', 'RandomForestClassifier_Best_Model_Params', 'AdaBoostClassifier_Best_Model_Params',
                     'XGBClassifier_Best_Model_Params'], axis=1, inplace=True)
```

```python
training_log_df
```

```python
training_log_mean = training_log_df.groupby(['Patient Subset', 'Feature Subset', 'Feature Selection', 'Engaged', 'Summarized', 'Vectorization']).mean() # , 'Model'
```

```python
training_log_std = training_log_df.groupby(['Patient Subset', 'Feature Subset', 'Feature Selection', 'Engaged', 'Summarized', 'Vectorization']).std() # , 'Model'
```

```python
# Create a csv file with two sheets with above dataframes
with pd.ExcelWriter('training_log_cost_sensitive_learning_0514_mean_std.xlsx') as writer:
    training_log_mean.to_excel(writer, sheet_name='Mean')
    training_log_std.to_excel(writer, sheet_name='Std')
```
